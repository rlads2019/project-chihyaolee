```python
f = open("247138_comma.txt", "r")
commas = f.read()
urls = commas.split(", ")
f.close()
```

```python
import requests
from requests_html import HTML
from six.moves import urllib
from random import randint
from time import sleep
```

```python
for url in urls[98857:148284]:
    resp = requests.get('https://www.ptt.cc/bbs/Gossiping/'+url+'.html', cookies={'over18': '1'})
    html = HTML(html=resp.text)
    try:
        title = html.find('#main-content > div:nth-child(3) > span.article-meta-value')[0].text
        if 'Re:' not in title:
            content = html.find('#main-container')[0].text
            f=open("ptt3.txt", "a+")
            f.write(content.replace('\n', '!NL!LA!')+'\n')
            f.close()
            print(url)
    except IndexError:
        title = False
    rest = randint(1, 2)
    sleep(rest)
    
```
